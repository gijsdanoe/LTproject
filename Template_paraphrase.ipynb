```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Vamsi/T5_Paraphrase_Paws")

model = AutoModelForSeq2SeqLM.from_pretrained("Vamsi/T5_Paraphrase_Paws")
```

```python
from transformers import pipeline
```

```python
text2text = pipeline('text2text-generation', model = "Vamsi/T5_Paraphrase_Paws")
text2text("paraphrase: This is something which I cannt understand at all.")
```

```python
text2text("paraphrase: [X] is a [MASK] frame.")
```

```python
text2text("paraphrase: [X] The argument frame of [X] is [MASK].")
```

```python
text2text("paraphrase: The argument frame of [X] is [MASK1]. Is it correct?")
```
